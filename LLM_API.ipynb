# Youtube API

```python
from youtube_transcript_api import YouTubeTranscriptApi
from urllib.parse import urlparse
from urllib.parse import parse_qs


def extract_text(youtube_link):
  video_text = YouTubeTranscriptApi.get_transcript(youtube_link)
  text=""
  for i in video_text:
    text+=f"{i['text']} "
  return text

def get_id(link):
  parsed_url = urlparse(link)
  captured_value = parse_qs(parsed_url.query)['v'][0]
  return captured_value
```

```python
from llama_index.core import VectorStoreIndex,SimpleDirectoryReader,ServiceContext
from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.core.prompts.prompts import SimpleInputPrompt

from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from llama_index.core import ServiceContext
from llama_index.embeddings.langchain import LangchainEmbedding
from llama_index.core.schema import TextNode
```

```python
system_prompt="""
You are a Q&A assistant. Your should answer questions as accurately as possible based on the instructions and context provided.
"""
query_wrapper_prompt=SimpleInputPrompt("<|USER|>{query_str}<|ASSISTANT|>")
```

```python
import torch


def split_text_into_nodes(temp_source, chunk_size=3000):
    nodes = []
    num_chunks = len(temp_source) // chunk_size + (1 if len(temp_source) % chunk_size != 0 else 0)
    for i in range(num_chunks):
        start_index = i * chunk_size
        end_index = min((i + 1) * chunk_size, len(temp_source))
        chunk_text = temp_source[start_index:end_index]
        node = TextNode(text=chunk_text, id=str(i + 1))
        nodes.append(node)
    return nodes


class LLM:
  def __init__(self):
    self.llm = HuggingFaceLLM(
        context_window=4096,
        max_new_tokens=256,
        generate_kwargs={"temperature": 0.0, "do_sample": False},
        system_prompt=system_prompt,
        query_wrapper_prompt=query_wrapper_prompt,
        tokenizer_name="meta-llama/Llama-2-7b-chat-hf",
        model_name="meta-llama/Llama-2-7b-chat-hf",
        device_map="auto",
        model_kwargs={"torch_dtype": torch.float16 , "load_in_8bit":True}
    )


    self.embed_model=LangchainEmbedding(
        HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2"))

    self.service_context=ServiceContext.from_defaults(
        chunk_size=1024,
        llm=self.llm,
        embed_model=self.embed_model
    )

  def prepare_model(self, text):
    self.nodes = split_text_into_nodes(text)
    print(len(self.nodes))
    self.index=VectorStoreIndex(self.nodes, service_context=self.service_context)
    self.query_engine=self.index.as_query_engine()


  def ask(self, question):
    response=self.query_engine.query(question)
    return response

```

```python
obj = LLM()
```

```python
from fastapi import FastAPI, Request
import nest_asyncio
import public_ip
```

```python
from fastapi.responses import JSONResponse
from fastapi.encoders import jsonable_encoder
```

```python
app = FastAPI()
```

```python
@app.post("/response")
async def get_link(request: Request):
  res = await request.json()
  print(res)
  question = res["messages"][0]["content"]
  return JSONResponse(jsonable_encoder({"assistant": obj.ask(question).response}))


@app.post("/makemodel")
async def prepare_model(request: Request):
  res = await request.json()
  print(res)
  youtube_link = res["messages"][0]["content"]
  print("YOUTUBE : ", youtube_link)
  if youtube_link != "":
    source_text = extract_text(youtube_link)
    obj.prepare_model(source_text)
  return JSONResponse(jsonable_encoder({"assistant": "SUCCESS"}))
```
